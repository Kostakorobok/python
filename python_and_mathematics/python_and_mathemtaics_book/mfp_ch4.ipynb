# Chapter 4: Calculus

```python
import sympy as sym
```

```python
x = sym.Symbol("x")
a = sym.Symbol("a")
b = sym.Symbol("b")
expression = (24 * x * (a - 4 * x) + 2 * (a - 8 * x) * (b - 4 * x)) / ((b - 4 * x)**4)
expression
```

```python
derivative = sym.diff(expression, x)
derivative
```

```python
first_equation = sym.Eq(derivative.subs({x: 0}), 0)
first_equation
```
