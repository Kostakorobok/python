Solve the following:

$\large \text{(a)} \; x + 3 = -1$

```python
x = sympy.Symbol("x")
expression = x + 3
expression
```

```python
expression = sympy.Eq(lhs = x + 3, rhs = -1)
expression
```

```python
sympy.solveset(expression, x)
```

$\large \text{(b)} \; 3x^2+2x=-5$

```python
expression = sympy.Eq(3 * x**2 + 2 * x, -5)
expression
```

```python
sympy.solveset(expression, x)
```
