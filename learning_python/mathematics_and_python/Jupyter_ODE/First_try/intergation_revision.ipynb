```python
from sympy import symbols, sin, cos, pi, integrate
import math
```

Example 8.43:
a) $$ \int_a^b f(x)\ dx = \int_1^2 (x^{4}+6x^{2}-4)\ dx$$

```python
x = symbols("x")
a, b, c = 1, 6, 4
equation_a = a*x**4 + b*x**2 - c
```

```python
result = integrate(equation_a, (x, 1, 2))
print(result)
```

Example 8.43 b)
$$ \int_a^b f(x) \ dx = \int_1^2 \frac{(x^{2} - 1)^{2}}{x^{2}} \ dx $$

```python
equation_b = ((x**2-1)**2)/(x**2)
result = integrate(equation_b, (x, 1, 2))
print(result)
```

Euler's number: 
$$e^{x}$$
Example c):
$$ \int_{-2}^{4} f(x) \ dx = \int_{-2}^{4} 4 \times e^{x} \ dx $$

```python
eulers = math.e**x
equation_c = 4 * eulers
result = integrate(equation_c, (x,-2, 4))
print(result)
```

Example d):
$$ \int_{0}^{\frac{\pi}{6}} (cos(3x)+2sin(3x)) \ dx $$

```python
equation_d = cos(3*x) + 2*sin(3*x)
result = integrate(equation_d, (x, 0, pi/6))
print(result)
```
