## Solving trigonometric functions

```python
# We want to solve the equation 4*tan(x)^2+3==15 for x in [0,2pi]
# The solutions are pi/3, 2*pi/3, 4*pi/3, and 5*pi/3=-pi/3

solve(4*(tan(x))^2+3==15, x, to_poly_solve=True) 
# This doesn't give all the solutions. You need to generate the other solutions
```

```python
# We want to solve the equation 4*tan(x)^2+3==15 for xin [0,2pi]
# If we change the solver, then we get all the solutions but in a less readable format   

solve(4*tan(x)^2+3==15, x, algorithm='sympy') 
```

```python
#To find a numeric approximation of the solution we can use the find_root() command. 
# This command will only return one solution on the specified interval, if one exists. 
# It will not find the complete solution set over the entire real numbers. 
# To find a complete set of solutions, you  must use find_root() repeatedly over cleverly selected intervals.

find_root(4*tan(x)^2+3==15, 0, 2*pi) 
```

```python
# Projections. We write a python function to produce the projection of a vector u onto a vector d 

def proj(u,d):
    p=u.dot_product(v)/(norm(d)^2)*d
    return p

u=vector([2,-1,0]); v=vector([3,1,-1]); w=proj(u,v); w # w is the projection of u onto v
```

```python
# Cross product of two vectors in R^3

u.cross_product(v)
```

```python
# We compute the volume of the parallelepiped determined by the three vectors u, v, and w in R^3.
# We write a python function to produce this volume.  

def vol(u,v,w):
    temp=abs(w.dot_product(u.cross_product(v)))
    return temp
    
u=vector([2,1,1]); v=vector([1,0,2]); w= vector([2,1,-1])

vol(u,v,w)
```

```python
# We are tossing one fair coin. Possible outcomes are 'head'=0 or 'tail'=1 

# We define the sample space and the probabilities. The sample space needs to be encoded 
# so that the elements go from 0 to number of elements in the sample space minus one  
# The number of elements in the sample space is given by the number of probabilities
# in P.

P=[0.5, 0.5] # This means that sample_space=[0,1]

# We create a discrete probability distribution.

X=GeneralDiscreteDistribution(P)
```

```python
# We get a random element from the sample space 
X.get_random_element()
```

### Binomial distribution

```python
# The binomial function calculates the number of ways to choose k successes in n trials.

binomial(6,2) #Say n=6, k=2
```

```python
# To define the binomial distribution we need three parameters n, p, and k.
# p represents the probability of a single success and 1-p is the probability of a single failure.
# n represents the number of independent trials, and k represents number of successes

# First we define a binomial random variable X2. To do so we pass n=9 and p=0.35.

X2=stats.binom(9,0.35)
```

```python
# We can compute the probability mass function (pmf) of X2 at a value k, namely P(k successes in n trials).
# This function is binomial(n, k) * p^k * (1-p)^(n-k)

X2.pmf(2) # Say k=2
```

```python
# We can compute the cumulative distribution function (cdf) of X1 at a value x, 
# namely P(X2<=k) of X2 at a value k=2, namely P(AT MOST k successes in n trials)
# P(AT MOST k successes in n trials)=P(0 successes in n trials)+...+P(k successes in n trials)

X2.cdf(5)
```

```python
X2.pmf(0)+X2.pmf(1)+X2.pmf(2)+X2.pmf(3)+X2.pmf(4)+X2.pmf(5)
```

```python
# We can also compute standard deviation, variance, mean, median, expected value of X2

X2.std(), X2.var(), X2.mean(), X2.median(), X2.expect()
```

### Pisson distribution

```python
# To define the Poisson distribution we need two parameters lambda and k.
# lambda represents therate for a Poisson distribution, namely 
# the average number of occurrences in a mostly-fixed population per unit of time
# k represents the number of rare events

# First we define a Poisson random variable X3. To do so we pass lambda=2/7.

X3=stats.poisson(2/7);
```

```python
# We can compute the probability mass function (pmf) of X3 at a value k, namely P(observe k rare events).
# This function is lambda^k * exp(-lambda)/factorial(k)

X3.pmf(3) #k=3
```

```python
# We can also compute standard deviation, variance, mean, median, expected value of X3

X3.std(), X3.var(), X3.mean(), X3.median(), X3.expect()
```
